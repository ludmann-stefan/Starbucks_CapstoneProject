```python
def get_top_customers (n):
    value_person = transcript_a.groupby (['person'])['count'].sum ().sort_values(ascending = False)
    value_person = pd.DataFrame (value_person.iloc [0:n]).rename (columns = {'count': 'visits'})

    return value_person

def get_user_interactions (user_id):
    user_int = transcript_a [transcript_a ['person'] == user_id] 
    return user_int

def get_offer_response (i):
    
    
    # data = 3 Parts... + transactions
    received_index = i
    received_matrix = received.loc [i]
    offer_id = (received.loc [i] ['offer_id']) 
    person = (received.loc [i] ['person'])
    received_time = received.loc [i] ['time']
    expiration_time = received.loc [i] ['expiration']    
    
    viewed_yn = []
    viewtoexpire = []
    completed_yn = []
    completed_yn_unknown = []
    compacttime = []
    comp_matrix = []
    liste = []
    

    
    view_matrix = viewed [viewed ['person'] == person][viewed [viewed ['person'] == person]['offer_id'] == offer_id]
    view_matrix = view_matrix [view_matrix ['time'] >= received_time]
    view_matrix = view_matrix [view_matrix ['time'] <= expiration_time]
    
    try: 
        viewtime = view_matrix.iloc [0]['time']
    except:
        viewtime = -98

        
    comp_matrix = completed [completed ['person'] == person][completed [completed ['person'] == person]['offer_id']== offer_id]
    comp_matrix = comp_matrix [comp_matrix ['time'] >= received_time]    
    
    try:
        completed_time = (comp_matrix.iloc [0]['time'])
    except: 
        completed_time = -99
    
    
    
    if viewtime >= 0:
        viewed_yn = 1
        viewtoexpire = expiration_time - viewtime
    else:
        viewed_yn = 0
        viewtoexpire = 0
            
    if completed_time >= 0:
        if viewtime >= 0:
            compacttime =  completed_time - viewtime
            completed_yn = 1
            completed_yn_unknown = 0
        else:
            compacttime = 0
            completed_yn_unknown =  1
            completed_yn = 0            
    else:
        completed_yn_unknown = 0
        compacttime = 0         
        completed_yn = 0
       
    
    index_received = received_index
    liste = index_received
    
    index_viewed = -1
    try: 
        index_viewed = view_matrix.index [0].astype (int)
        liste.append (index_viewed)
    except: index_viewed = index_viewed

    index_completed = -1
    try:
        index_completed = comp_matrix.index [0].astype (int)
        liste.append (index_completed)
    except: index_completed = index_completed


    start = viewtime
    end = min (completed_time, expiration_time)
            
    
    visits = transactions [transactions ['person'] == person][transactions [transactions ['person'] == person] ['time'] >= start]        
    visits = visits [visits ['time'] < end]
    visits_count = len (visits [visits ['time'] < end])

    
    
    vector = ['offer_id', 'person', 'viewed_yn', 'viewtoexpire', 'completed_yn', 'compacttime', 'completed_yn_unknown', 'index_received', 'index_viewed', 'index_vompleted', 'visits_count']
    
    matrix = pd.DataFrame ([offer_id, person, viewed_yn, viewtoexpire, completed_yn, compacttime, completed_yn_unknown, index_received, index_viewed, index_completed, visits_count])
    
    return matrix



### refer to the transaction that completes the offer!!!

# something like 'index -1' in the working_transcript id


### refer to transactions after seeing the offer, that do not complete the offer

# something like transactions of person between viewtime and expiration... save the indecees in a column/ array
# but stop if offer completed!


# number of visits till successfull
# matrix of cols offers and rows persons 
# if more often the same offer sent then use the average


def get_user_interaction (n):
    probants = transcript['person'].value_counts().index
    
    day_nr = (transcript [transcript ['person'] == probants [n]]['time']/24)
    interaction = transcript [transcript ['person'] == probants [n]].merge (pd.DataFrame (day_nr.astype (int)), left_index = True, right_index = True)
    
    return (interaction)


def get_userint_dayPerWeek (n):
    total_user_int  = get_user_interaction (n) [get_user_interaction (n) ['event'] == 'transaction']

    day_nr = 0
    total = []
    while day_nr < max (total_user_int['time_y']):
        
    
        day_arr = []
        i = 0
        while i < len(total_user_int [total_user_int['time_y'] == day_nr]):
            day_arr.append (total_user_int [total_user_int['time_y'] == day_nr].iloc [i]['value'].get('amount'))
            i+=1
        
        total.append (round (sum(day_arr), 2))
        day_nr += 1
    while day_nr < 35:
        total.append (0)
        day_nr += 1

    week1, week2, week3, week4, week5 = total [0:7], total [7:14], total [14:21], total [21:28], total [28:35]
    
    month = pd.DataFrame (week1)
    month [1] = week2
    month [2] = week3
    month [3] = week4
    month [4] = week5
    month ['sum'] = month.sum (axis = 1)
    month = month.T
    
    
    return (month)

def get_offer_receive_time (n):
    offers = get_user_interaction (n) [get_user_interaction (n)['event'] == 'offer received']
    i = 0
    value_id = []
    while i < len (offers):
        value_id.append (offers.iloc [i]['value'].get ('offer id'))
        i += 1
    offers ['value_id'] = value_id
    
    viewed = get_user_interaction (n) [get_user_interaction (n)['event'] == 'offer viewed']
    i = 0
    value_id = []
    while i < len (viewed):
        value_id.append (viewed.iloc [i]['value'].get ('offer id'))
        i += 1
    viewed ['value_id'] = value_id
    
    completed = get_user_interaction (n) [get_user_interaction (n)['event'] == 'offer completed']
    i = 0
    value_id = []
    while i < len (completed):
        value_id.append (completed.iloc [i]['value'].get ('offer_id'))
        i += 1
    completed ['value_id'] = value_id    
    reward = []
    i = 0
    while i < len (completed):
        reward.append (completed.iloc [i]['value'].get ('reward'))
        i += 1
    completed ['reward'] = reward  
    
    
    return (offers, viewed, completed)

def get_response (n):
    offers, viewed, completed = get_offer_receive_time (n)
    
    observation = []
    i = 0
    while i < len (offers):
        
        watched, initial, received = [], [], []
        
        id_offer = offers['value_id'].iloc [i]
        
        initial = int (offers.iloc [i] ['time_x'])

        if len (viewed [viewed ['value_id'] == id_offer] ['time_x']) > 0:
            watched = int (viewed [viewed ['value_id'] == id_offer].iloc [0] ['time_x'])
        else: watched = 0
        
        if len (completed [completed ['value_id'] == id_offer] ['time_x']) > 0:
            received = int (completed [completed ['value_id'] == id_offer].iloc [0]  ['time_x'])
        else: received = 0
            
        realiation = watched - initial
        reaction = received - watched
        if len (viewed [viewed ['value_id'] == id_offer] ['time_x']) == 0:
            realiation = 0
            reaction = 0
        if len (completed [completed ['value_id'] == id_offer] ['time_x']) == 0:
            reaction = 0
        
        observation.append ([ offers.iloc [i] ['value_id'], realiation, reaction])
        
        i += 1
    return (observation)

def user_overview (n):
    dayly = get_userint_dayPerWeek (n)
    response = get_response (n)
    interactions = get_user_interaction (n)
    
    response = pd.DataFrame (response).merge (portfolio, left_on = [0], right_on = ['id'])
    
    return (dayly, response, interactions)

```

```python
#Hopkins Statistic is a way of measuring the cluster tendency of a data set.
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
 
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) # heuristic from article [1]
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
#Principal Component
from sklearn.decomposition import PCA

pca = PCA(n_components=3, whiten=True)

Num_features=Data_pca.select_dtypes(include=[np.number]).columns

x=Data_pca[Num_features]
principalComponents = pca.fit_transform(x)

# Cumulative Explained Variance
cum_explained_var = []
for i in range(0, len(pca.explained_variance_ratio_)):
    if i == 0:
        cum_explained_var.append(pca.explained_variance_ratio_[i])
    else:
        cum_explained_var.append(pca.explained_variance_ratio_[i] + 
                                 cum_explained_var[i-1])

print(cum_explained_var)
```
